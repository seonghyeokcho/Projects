### 지하철 구별 분포도

```python
subway = pd.read_csv('data/서울시지하철정보.csv', encoding='cp949')
subway
```

```python
subway = subway.query("호선 not in ['E','I','I2','KK','SU','U']")
subway
```

```python
UI = subway.query("호선 == 'UI'")
UI = UI[['전철역명','전철역명_영문','호선','주소','X좌표(WGS)','Y좌표(WGS)']]
UI = UI.drop_duplicates()
UI = UI.query("전철역명 not in ['보문','성신여대입구','신설동']")
UI
```

```python
# 우이신설 노선 주소 누락 추가
idx = UI.index
addr = ['서울 강북구 우이동 72-223', '서울 강북구 수유동 338-47', '서울 성북구 보국문로 지하 88', '서울 강북구 우이동 16-27',
         '서울 강북구 미아동 791-4755', '서울 강북구 미아동 745-149', '서울 강북구 우이동 57-38', '서울 강북구 솔샘로 지하 189',
         '서울 성북구 정릉동 966-330', '서울 강북구 수유동 472-691']
for i in range(len(UI)):
    subway.at[idx[i], '주소'] = addr[i]
subway
```

```python
UI = subway.query("호선 == 'UI'")
UI = UI[['전철역명','전철역명_영문','호선','주소','X좌표(WGS)','Y좌표(WGS)']]
UI = UI.drop_duplicates()
UI = UI.query("전철역명 not in ['보문','성신여대입구','신설동']")
UI
```

```python
import requests; from urllib.parse import urlparse; import json

subway_lat = []
subway_lng = []
subway_addr = UI.주소
for addr in subway_addr:
    url = 'https://dapi.kakao.com/v2/local/search/address.json?query=' + addr
    headers = {"Authorization": "KakaoAK 7ff520ce7fc46439d58b8c4fc9eb0c7c"}
    result = json.loads(str(requests.get(url, headers=headers).text))
    match_first = result['documents'][0]['address']
    subway_lat.append(float(match_first['y']))
    subway_lng.append(float(match_first['x']))
idx = UI.index
for i in range(len(UI)):
    subway.at[idx[i], 'X좌표(WGS)'] = subway_lat[i]
    subway.at[idx[i], 'Y좌표(WGS)'] = subway_lng[i]
subway
```

```python
subway = subway[['전철역명','전철역명_영문','주소','X좌표(WGS)','Y좌표(WGS)']]
subway = subway.rename(columns={'X좌표(WGS)':'lat','Y좌표(WGS)':'lng'})
subway
```

```python
subway = subway[subway.주소.notnull()]
subway
```

```python
subway['city'] = subway['주소'].dropna().apply(lambda x: x.split(' ')[0])
subway['gu'] = subway['주소'].dropna().apply(lambda x: x.split(' ')[1])
subway
```

```python
subway = subway.query("city in ['서울특별시','서울']")
subway
```

```python
subway = subway.drop_duplicates('lat' and 'lng')
subway
```

```python
print(subway.전철역명.nunique(),subway.전철역명_영문.nunique(),subway.주소.nunique(),
      subway.lat.nunique(),subway.lng.nunique(),subway.city.nunique(),subway.gu.nunique())
```

```python
subway_cnt = subway.groupby('gu')['전철역명'].agg('count').reset_index()
subway_cnt
```

```python
subway_cnt.columns = ['gu','subway_cnt']
subway_cnt.head()
```

```python
subway_cnt.to_csv('data/Processeddata/subway_cnt(UI).csv', index=False, encoding='cp949')
```

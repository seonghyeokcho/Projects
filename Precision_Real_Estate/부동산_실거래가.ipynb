```python
import googlemaps
gmaps = googlemaps.Client(key='AIzaSyDV5heLcWZDRW-8Hggo5O64TqmnSQqCT-0')

apt_addr = real_estate.동 + " " + real_estate.도로명 + " " + real_estate.단지명
idx = real_estate.index

for i in range(len(real_estate)):
    apt_lat = []
    apt_lng = []
    for addr in apt_addr:
        tmp = gmaps.geocode(addr, language='ko')
        if len(tmp) == 0:
            apt_lat.append(0)
            apt_lng.append(0)
            continue
        apt_lat.append(tmp[0].get('geometry')['location']['lat'])
        apt_lng.append(tmp[0].get('geometry')['location']['lng'])
    real_estate.at[idx[i], 'lat'] = apt_lat[i]
    real_estate.at[idx[i], 'lng'] = apt_lng[i]
```
